### Data Extraction

```python
# Dependencies and packages
import os
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import math as math
import datetime as dt
import seaborn as sns
import pandas as pd
from sodapy import Socrata
```

```python
client = Socrata("data.austintexas.gov", None)

results = client.get("i26j-ai4z", limit=850000)

# Convert to pandas DataFrame
results_df = pd.DataFrame.from_records(results)
results_df
```

```python
#df = pd.read_csv('Data_Files/311_Unified_Data_Test_Site_2019.csv', low_memory=False)
#df_backup = df
```

```python
results_df = results_df[results_df["sr_type_code"]== "DOCKMOBI"]
```

### Data Cleaning

```python
# Renaming Column Names
results_df.columns = results_df.columns.str.replace('_',' ').str.title().str.replace('Sr','SR')
results_df
```

```python
results_df.count()
```

```python
# Delete unwanted column names
clean_df = results_df[results_df.columns[~results_df.columns.str.contains(':@Computed Region')]]
list(clean_df.columns)
```

```python
shared_mobility_calls = clean_df[clean_df["SR Type Desc"] == "Shared Micromobility"]
shared_mobility_calls
```

```python
zipcodes = shared_mobility_calls["SR Location Zip Code"].value_counts()
zipcodes
```
